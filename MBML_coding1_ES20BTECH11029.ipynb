```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 1

```python
N = 1000
eta = 0.02
```

### Part (a)

$$
\begin{align}
f(x) &= x^TAx + b \\
\nabla f(x) &= 2Ax
\end{align}
$$

```python
A = np.array([[2, -1, -1],
              [-1, 2, 0],
              [-1, 0, 1]], dtype=np.float32)

b = np.array(1, dtype=np.float32)

print(f"A = {A}, \nb = {b}")
```

```python
f = lambda x: np.dot(np.dot(x.T, A), x) + b
```

#### Analytic Solution

$$
\begin{align}
\nabla f(x) &= 0 \\
2Ax &= 0 \\
x &= A^{-1}0 = 0
\end{align}
$$

```python
x = np.array([[0], [0], [0]], dtype=np.float32)
print(f"x_min using Analytic method = {x}")
print(f"f(x_min) = {f(x)[0][0]}")
```

#### Gradient Descent

```python
# Gradient Descent
grad_f = lambda x: 2 * np.dot(A, x)
```

```python
# Initial x
x = np.random.rand(3, 1)
```

```python
for k in range(1, N+1):
    grad = grad_f(x)
    x = x - eta * grad
    print(f"Iteration: {k}")
    print(f"xk = {x}")
    print(f"Gradient = {grad}")
    print(f"f(xk) = {f(x)[0][0]}")
```

```python
print(f"x_min using Gradient Descent = {x}")
print(f"f(x_min) = {f(x)[0][0]}")
```

### Part (b)

$$
f(x) = \|Ax - b\|^2 \\
\nabla f(x) = 2 A^T (Ax - b)
$$

```python
A = np.array([[1, 2],
              [2, 4],
              [3, 1]], dtype=np.float32)

b = np.array([[1], [3], [1]], dtype=np.float32)

print(f"A = {A}, \nb = {b}")
```

```python
f = lambda x: np.square(np.linalg.norm(np.dot(A, x) - b))
```

#### Analytic Solution

$$
\begin{align}
\begin{split}
\nabla f(x) &= 0 = 2A^T (Ax - b) \\
(Ax - b) &= 0 \\
x &= A^{-1}b
\end{split}
\end{align}
$$

```python
x = np.linalg.lstsq(A, b)[0]
print(f"x_min using Analytic Method = {x}")
print(f"f(x_min) = {f(x)}")
```

#### Gradient Descent

```python
grad_f = lambda x: 2 * np.dot(A.T, (np.dot(A, x) - b))
```

```python
x = np.random.rand(2, 1)
```

```python
for k in range(1, N+1):
    grad = grad_f(x)
    x = x - eta * grad
    print(f"Iteration: {k}")
    print(f"xk = {x}")
    print(f"Gradient = {grad}")
    print(f"f(xk) = {f(x)}")
```

```python
print(f"x_min using Gradient Descent= {x}")
print(f"f(x_min) = {f(x)}")
```

### Part (c)

f(x) &= \|Ax - b\|^2 \\

```python
f = lambda x: np.square(np.linalg.norm(np.dot(A, x) - b))
```

#### Analytic Solution

$$
\begin{align}
\nabla f(x) &= 0 = 2A^T (Ax - b) \\
(Ax - b) &= 0 \\
x &= A^{-1}b \\
\end{align}

```python
x = np.linalg.lstsq(A, b)[0]
print(f"x_min using Analytic Method = {x}")
print(f"f(x_min) = {f(x)}")
```

```python
grad_f = lambda x: 2 * np.dot(A.T, (np.dot(A, x) - b))
```

```python
for k in range(1, N+1):
    grad = grad_f(x)
    x = x - 0.002 * grad
    print(f"Iteration: {k}")
    print(f"xk = {x}")
    print(f"Gradient = {grad}")
    print(f"f(xk) = {f(x)}")
```

## Question 2

$$
\begin{align}
f(x) &= x^TAx - 2b^Tx + c\\
\nabla f(x) &= 2(Ax - b)
\end{align}
$$

```python
from sklearn.datasets import make_spd_matrix
A = make_spd_matrix(10)
b = np.random.rand(10, 1)
c = np.random.rand()

print(f"A = {A}, \nb = {b}, \nc = {c}")
```

```python
f = lambda x: np.dot(np.dot(x.T, A), x) - 2 * np.dot(b.T, x) + c
```

#### Part (a): Analytic Solution

$$
\begin{align}
\nabla f(x) &= 0 = 2 (Ax - b) \\
Ax - b &= 0 \\
\end{align}
$$

```python
x = np.linalg.solve(A, b)
print(f"x_min using Analytic Solution= {x}")

f_min_analytic = f(x)[0][0]
print(f"f(x_min) = {f_min_analytic}")
```

#### Part (b): Gradient Descent

```python
x = np.full(10, 1/10, dtype=np.float32)
x
```

```python
l_gd = 1 / (2 * np.linalg.norm(A, ord=2) + np.linalg.norm(b))
l_gd
```

```python
grad_f = lambda x: 2 * (np.dot(A, x) - b)
```

```python
y = [f(x)]
iters = [0]

for k in range(1, 1001):
    grad = grad_f(x)
    x = x - l_gd * grad
    y.append(f(x)[0][0])
    iters.append(k)
```

```python
# Create the plot
plt.figure(figsize=(8, 6))

plt.plot(iters, y, linewidth=2, color='blue')

plt.axhline(y=f_min_analytic, color='red', linestyle='-', label=f'f(x) using analytic solution = {f_min_analytic}')

for i, (xi, yi) in enumerate(zip(iters, y)):
    if xi % 200 == 0:
        plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

plt.xlabel('Iteration number')
plt.ylabel('f(x)')
plt.legend()

plt.title('Minimization of f(x) using Gradient Descent vs Analytic Method')
plt.show()
```
