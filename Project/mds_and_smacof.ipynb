```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.manifold import MDS

# Default parameters
FIGURESIZE=(10, 8)
FONTSIZE=24

plt.rcParams['figure.figsize'] = FIGURESIZE
plt.rcParams['font.size'] = FONTSIZE

plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Swiss Roll

```python
t = np.arange(0, 10, .01)
x0 = np.cos(t) * (t + 1)
x1 = np.sin(t) * (t + 1)
x2 = [20 * random.random() for x in x0]

x0 = np.array([x0])
x1 = np.array([x1])
x2 = np.array([x2])
X = np.concatenate([x0, x1, x2], axis=0).T

X.shape
```

```python
# 2D Representation of dataset
fig = plt.figure(figsize=FIGURESIZE)
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=t, cmap=plt.cm.rainbow)
```

```python
fig = plt.figure(figsize=FIGURESIZE)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.rainbow)
```

```python
# MDS
mds = MDS(n_components=2, n_jobs=-1)
X_transform = mds.fit_transform(X)
```

```python
fig = plt.figure(figsize=FIGURESIZE)
ax = fig.add_subplot(111)
ax.scatter(X_transform[:, 0], X_transform[:, 1], c=t, cmap=plt.cm.rainbow)
```

```python
# Smacof
stress = mds.stress_
print(stress)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
dist_euclidean = euclidean_distances(X)
mds = MDS(dissimilarity='precomputed', random_state=0)

# Get the embeddings
X_transform_L1 = mds.fit_transform(dist_euclidean)
```

```python
X_transform_L1
```

```python
fig = plt.figure(figsize=FIGURESIZE)
ax = fig.add_subplot(111)
ax.scatter(X_transform_L1[:, 0], X_transform_L1[:, 1], c=t, cmap=plt.cm.rainbow)
```

# Arrow

```python
t = np.arange(-1, 1, 0.001)
x0 = t
x1 = [0.02 * random.uniform(-1, 1) if (v < 0.5 and v > -0.5)
      else 2 * (1 + v) * random.uniform(-1, 1) if v <= -0.5
      else 2 * (-0.5 + v) * random.uniform(-1, 1) for v in t]
x2 = np.cos(t * 5 + 1.5)
x0 = np.array([x0])
x1 = np.array([x1])
x2 = np.array([x2])

X = np.concatenate([x0, x1, x2], axis=0).T
X.shape
```

```python
fig = plt.figure(figsize=FIGURESIZE)
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=t, cmap=plt.cm.rainbow)
```

```python
fig = plt.figure(figsize=FIGURESIZE)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.rainbow)
```

```python
# MDS
mds2 = MDS(n_components=2, n_jobs=-1)
X_transform2 = mds2.fit_transform(X)
```

```python
fig = plt.figure(figsize=FIGURESIZE)
ax = fig.add_subplot(111)
ax.scatter(X_transform2[:, 0], X_transform2[:, 1], c=t, cmap=plt.cm.rainbow)
```

```python
# Smacof
stress2 = mds2.stress_
print(stress2)
```
